```python
### NIVEL 1 ###

# Desde un Jupyter Notebook realizarás los siguientes ejercicios utilizando la librería requests de Python.

# Exploración básica con una API de laboratorio.

# 1. Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:

# Lista de publicaciones (/posts)

import requests

url_posts = "https://jsonplaceholder.typicode.com/posts"

respuesta_posts = requests.get(url_posts)

posts = respuesta_posts.json()

print(posts)

```

```python
# Lista de usuarios (/users)

url_users = "https://jsonplaceholder.typicode.com/users"

respuesta_users = requests.get(url_users)

users = respuesta_users.json()

print(users)
```

```python
# Lista de tareas (/todos)

url_todos = "https://jsonplaceholder.typicode.com/todos"

respuesta_todos = requests.get(url_todos)

todos = respuesta_todos.json()

print(todos)
```

```python
# 2. Muestra por pantalla:

# La cantidad total de cada recurso.

print("Cantidad de posts:", len(posts))

print("Cantidad de usuarios:", len(users))

print("Cantidad de tareas:", len(todos))
```

```python
# El código de estado de cada petición.

print('El codigo de estado de la peticion es:', respuesta_posts.status_code)

print('El codigo de estado de la peticion es:', respuesta_users.status_code)

print('El codigo de estado de la peticion es:', respuesta_todos.status_code)
```

```python
# 4. Realiza una petición POST para crear una nueva publicación ficticia. Incluye el título, el cuerpo del mensaje y un userId.

url = "https://jsonplaceholder.typicode.com/posts"

post_ficticio = {
    "title": "Publicacion de Lumi",
    "body": "Hola mundo, vamos a aprender Python con entusiasmo",
    "userid": "777"
}

respuesta = requests.post(url, json= post_ficticio)

# Muestra la respuesta JSON.

print("Respuesta JSON:", respuesta.json())

# Muestra el código de estado.

print("El codigo de estado es:", respuesta.status_code)
```

```python
# 5. Realiza una petición PATCH para modificar parcialmente una publicación existente.

url = "https://jsonplaceholder.typicode.com/posts/7"

url_modificado = {
    "title": "Titulo modificado"
}

respuesta = requests.patch(url, json=url_modificado) 

# Muestra la respuesta JSON.

print("Respuesta JSON:", respuesta.json()) 

# Muestra el código de estado.

print("El codigo de estado es:", respuesta.status_code)
```

```python
# 6. Haz una petición DELETE sobre una publicación.

url = "https://jsonplaceholder.typicode.com/posts/7"

respuesta = requests.delete(url) 

# Muestra la respuesta JSON.

print("Respuesta JSON:", respuesta.text) 

# Muestra el código de estado.

print("El codigo de estado es:", respuesta.status_code)
```

```python
### NIVEL 3 ###

# 1. Utiliza la API de Open Data BCN para:

# Buscar un dataset de tu interés mediante package_search y package_show. 

import requests

url_opendata = "https://opendata-ajuntament.barcelona.cat/data/api/3/action"

params = {
    "q": "poblacio", 
    "rows": 5   
}

response = requests.get(f"{url_opendata}/package_search", params=params) 
response.raise_for_status() 

data = response.json() 

datasets = data["result"]["results"] 

print(f"Datasets encontrados: {len(datasets)}") 

# hacemos un bucle para decidir que dataset nos interesa
for i, d in enumerate(datasets):
    print(f"{i+1}. {d['title']}")
    print(f"   ID: {d['id']}")
    print(f"   Nº recursos: {len(d['resources'])}")
```

```python
# elegimos el primer dataset encontrado
dataset_id = datasets[0]["id"]

response = requests.get(f"{url_opendata}/package_show", params={"id": dataset_id})
response.raise_for_status()

dataset_detail = response.json()["result"] 

print("Título del dataset:") 
print(dataset_detail["title"])

print("\nRecursos disponibles:")
for r in dataset_detail["resources"]:
    print(f"- Nombre: {r['name']}")
    print(f"  Formato: {r['format']}")
    print(f"  Resource ID: {r['id']}")
```

```python
# 2. De los resultados obtenidos, selecciona uno que tenga recursos disponibles en CSV o JSON.

datasets = data["result"]["results"]

# Vamos a ver qué datasets tienen recursos CSV o JSON
for d in datasets:
    for r in d["resources"]:
        if r["format"].upper() in ["CSV", "JSON"]:
            print("Dataset:", d["title"])
            print("Resource:", r["name"])
            print("Formato:", r["format"])
            print("Resource ID:", r["id"])
            print("-"*40)
```
