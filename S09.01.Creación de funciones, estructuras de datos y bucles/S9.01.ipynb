```python
# NIVEL 1
# Escribe una función que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deberá ingresar estos datos. 
# La función debe clasificar el resultado en sus respectivas categorías.

def indice_masa_corporal(peso, estatura):
    """Calcula el índice de masa corporal (IMC) a partir de datos dados por el ususario"""

    try:
        peso = float(peso)       # Convertir peso a numero
        estatura = float(estatura)  # Convertir estatura a numero
    except:
        return "Error: ingresa un número válido"  # Error si no es numerico
    
    # Validar rangos razonables
    if (peso <= 0 or estatura <= 0) or ((peso < 2 or peso > 400) or (estatura < 0.5 or estatura > 2.5)):
        return f'Error: valor imposible'
    
    # Calcular IMC
    IMC = round((peso/estatura**2), 2)  
    
    # Clasificar IMC
    if IMC < 18.5:
        return f'Su IMC es {IMC}, indica bajo peso'
    elif IMC < 25:
        return f'Su IMC es {IMC}, indica un peso normal'
    elif IMC < 30:
        return f'Su IMC es {IMC}, indica sobrepeso'
    else:
        return f'Su IMC es {IMC}, indica obesidad'
    
peso_usuario = input('Introduce tu peso en kg')     # Pedir peso
estatura_usuario = input('Introduce tu estatura en m')  # Pedir estatura

print(indice_masa_corporal(peso_usuario,estatura_usuario))  # Ejecutar función

 
```

```python
# Selecciona al menos 2 conversores, de modo que al introducir una temperatura devuelva, como mínimo, dos conversiones, 
# de modo que se puedan guardar (recuerda que un print() no se puede guardar nunca).


def conversor_temp(valor,unidad):
    """
    Convierte una temperatura dada, en cualquier unidad a todas las demas unidades: 
    en Celsius (C), Fahrenheit (F), Kelvin (K), Rankine (Ra) o Reaumur (Re)
    """
 
    # Validacion de valor numerico
    try:
        valor = float(valor)
    except ValueError:
        return 'Error: ingresa un numero valido.'

    # Convertimos todo a Celsius 
    c = None
    if unidad == "c":
        c = round(valor,2)
    elif unidad == "f":
        c = round((valor - 32) * 5/9,2)
    elif unidad == "k":
        c = round(valor - 273.15,2)
    elif unidad == "ra":
        c = round((valor - 491.67) * 5/9,2)
    elif unidad == "re":
        c = round(valor * 1.25,2)
    else:
        return "Error: unidad no reconocida. Usa 'C', 'F','K', 'Ra' o 'Re'."

    # Validacion de rango 
    if c < -90 or c > 60:
        return 'Error: valor imposible.'

    # Diccionario con todas las conversiones
    temp_convertidas = {
        "Celsius": c,
        "Fahrenheit": (c * 9/5) + 32,
        "Kelvin": c + 273.15,
        "Rankine": (c + 273.15) * 9/5,
        "Reaumur": c * 0.8
    }

    return temp_convertidas

valor_usuario = input('Introduzca temperatura: ') 
unidad_usuario = input("Unidad de temparatura (C, F, K, Ra o Re): ").lower()

# Llamada a la funcion
print(conversor_temp(valor_usuario,unidad_usuario))
```

```python
# Escribe una función que dado un texto, muestre las veces que aparece cada palabra. 
# Intenta que se gestionen todas las posibles casuísticas que hagan que el programa no funcione correctamente. 
# ¿Cuál es la longitud media de las palabras del texto que has escrito?

def conteo_palabras(texto):    
    """Calcula las veces que aparece cada palabra en un texto dado"""

    try:
        texto = str(texto)
    except:
        return 'Error: el texto no es valido '

    try:
        palabras = texto.split()
        if len(palabras) == 0:
            return 'Error: el texto esta vacio'
    except:
        return 'Error: el texto no se puede procesar'
    
    conteo ={}
    longitud_total = 0

    for palabra in palabras:
        palabra = palabra.lower().strip('.,!?;:()[]{}"\'') 
        longitud_total += len(palabra)
        if palabra in conteo:
            conteo[palabra] += 1
        else:
            conteo[palabra] = 1

    longitud_media = round(longitud_total/len(palabras),2)

    return conteo,{longitud_media}
    
texto_usuario = input('Escribe un texto o frase')
conteo, longitud_media = conteo_palabras(texto_usuario)
print(conteo)
print(f'La longitud media de las palabras del texto es {longitud_media}')
```

```python
# Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y los resultados los necesita a la inversa, es decir, intercambiando las claves y valores. 
# Los valores y claves en el diccionario original son únicos; si éste no es el caso, la función debería imprimir un mensaje de advertencia, 
# junto con una lista con los valores asociados a la clave repetida.

def diccionario_inverso(diccionario):
    """Recibe un diccionario dado y lo devuelve con sus valores y claves invertidos"""

    if  not isinstance (diccionario,(dict)):
        return 'Error: formato erroneo'
    
    valores_clave_repetida = {} 
    for clave, valor in diccionario.items(): 
        if valor not in valores_clave_repetida: valores_clave_repetida[valor] = [clave] 
        else: valores_clave_repetida[valor].append(clave)

    diccionario_invertido = {} 
    for clave,valor in valores_clave_repetida.items(): 
        if len(valor) >1: print(f'Error: el valor {valor} de la clave {clave} esta repetido') 
        else: diccionario_invertido[clave] = valor[0] 

    return diccionario_invertido

entrada = input("Introduce pares clave:valor separados por coma, ej. a:1,b:2: ")
diccionario_usuario = dict(item.split(":") for item in entrada.split(","))

print(diccionario_inverso(diccionario_usuario))
```

```python
# NIVEL 2
# El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas 
# habituales del diccionario según la letra con la que comienzan, es decir, las claves deben ir de la A a la Z y dentro de la A debemos ir de la A a Z.

def contador_palabras_archivo(nombre_archivo):
    """Recibe archivos txt y cuenta la frecuencia que aparece cada palabra y las ordena alfabeticamente"""

    if not isinstance (nombre_archivo,(str)):
        return 'Error: formato erroneo'
    
    try:
        with open (nombre_archivo,'r') as f:
            texto = f.read()
    except FileNotFoundError:
        return ('Error: el archivo no existe')

    if not nombre_archivo.endswith('.txt'):
        return'Error: el archivo no es txt'
    
    palabras = texto.split()
    conteo_palabras = {}

    for palabra in palabras:
        palabra = palabra.lower().strip('.,!?;:()[]{}"\'') 

        if not palabra:   
            continue

        if palabra in conteo_palabras:
            conteo_palabras[palabra] += 1
        else:
            conteo_palabras[palabra] = 1
    
    palabras_ordenadas = list(conteo_palabras)
    palabras_ordenadas.sort()
    palabras_orden_alfa = {}

    for palabra in palabras_ordenadas:
        palabras_orden_alfa[palabra] = conteo_palabras[palabra]

    grupos_palabras ={}
    for palabra in palabras_ordenadas:
        letra = palabra[0]

        if letra not in grupos_palabras:
            grupos_palabras[letra] = {}

        grupos_palabras[letra][palabra] = conteo_palabras[palabra]
        
    return grupos_palabras

print (contador_palabras_archivo(r"C:\Users\damia\OneDrive\Desktop\Otras cositas\Bootcamp IT Academy LOA\tu me quieres blanca.txt"))
              
```

```python
# El cliente recibe una lista de datos y necesita generar dos listas, la primera donde estarán todos los elementos que pudieron convertirse en 
# flotantes y la otra donde están los elementos que no pudieron convertirse. Ejemplo de la lista que recibe el cliente:

def convertor_float (lista):
    """Convierte una lista dada en dos listas (una que cuyos 
    elementos sean convirtieron en float a partir de la lista original y otra que no)"""

    if  not isinstance (lista,(list)):
        return 'Error: formato no valido'

    if len(lista) == 0:
        return 'Error: la lista esta vacia'


    lista_float = []
    lista_no_float = []

    def procesar_elemento(datos): #se encarga de procesar cualquier elemento de la lista
                                  #Esta función es recursiva para manejar listas o tuplas anidadas.  
        if isinstance(datos, (list, tuple)):    
            for dato in datos:                  # dato es el sub elemetno dentro de una lista o tupla
                procesar_elemento(dato)
        else:
            try:
                valor = float(datos)  # valor representa la conversión exitosa a float.
                lista_float.append(valor)
            except:
                lista_no_float.append(datos)

    for elemento in lista:
        procesar_elemento(elemento)  # Recorremos cada elemento de la lista que recibimos como parámetro.

    return lista_float, lista_no_float

print(convertor_float([ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]))
```
